## 🧩Paso 3: Resumen estadistico por emergencia ambiental


### 3.1 Agrupación por emergencia ambiental

🎯**Objetivo**: Resumir las variables clave de la tabla **emergencia ambiental**, creando métricas que representen su comportamiento real.



```python
# suma de emergencia del 1° al 3° trimestre de 2025
emergencia[['derrame', 'explosion', 'fuga', 'incendio', 'otro']].sum()
```

```python
# Emergencias por estado
emergencias_por_estado = (
    emergencia.groupby('estado')[['derrame', 'incendio', 'explosion','fuga','otro']]
    .sum()
)

emergencias_por_estado['total_emergencias'] = (
    emergencias_por_estado['derrame'] +
    emergencias_por_estado['incendio'] +
    emergencias_por_estado['explosion'] +
    emergencias_por_estado['fuga'] +
    emergencias_por_estado['otro']
)

emergencias_por_estado = emergencias_por_estado.sort_values(
    by='total_emergencias',
    ascending=False
)

emergencias_por_estado.head()
```

```python
# Estado con mas emergencias ambientales
estado_max = emergencias_por_estado['total_emergencias'].idxmax()

print(f"El estado con la mayor cantidad de emergencias es: {estado_max}")
```

```python
# Estado con menos emergencias ambientales
estado_min = emergencias_por_estado['total_emergencias'].idxmin()

print(f"El estado con la menor cantidad de emergencias es: {estado_min}")
```

```python

# Estado con mas derrames
estado_derrame = emergencias_por_estado['derrame'].idxmax()
print(f"El estado con la mayor cantidad de derrames es: {estado_derrame}")
```

```python
# Estado con mas incendios
estado_incendio = emergencias_por_estado['incendio'].idxmax()
print(f"El estado con la mayor cantidad de incendios es: {estado_incendio}")
```

```python
# Estado con mas explosiones
estado_explosion = emergencias_por_estado['explosion'].idxmax()
print(f"El estado con la mayor cantidad de explosiones es: {estado_explosion}")
```

```python
# Estado con mas fugas
estado_fugas = emergencias_por_estado['fuga'].idxmax()
print(f"El estado con la mayor cantidad de fugas es: {estado_fugas}")
```

```python
#indice de diversidad de emergencias por estado (el estado tiene muchos tipos de problemas o solo uno)
import numpy as np

def diversidad(row):
    p = row / row.sum()
    return -(p * np.log(p + 1e-9)).sum()

emergencias_por_estado['diversidad'] = emergencias_por_estado[['derrame','incendio','explosion','fuga','otro']].apply(diversidad, axis=1)
print(emergencias_por_estado)
```
